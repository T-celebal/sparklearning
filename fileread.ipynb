```
%python
df = spark.read.csv("/FileStore/tables/yellow_tripdata_2018_new.csv", header="true", inferSchema="true")             
display(df)             
```

```
%python
df_revenue = df.withColumn("Revenue", df.fare_amount+df.extra+df.mta_tax+df.tip_amount+df.tolls_amount+df.improvement_surcharge+df.total_amount)
```

```
display(df_revenue)
```

```
from pyspark.sql import functions as func

df.groupBy("PULocationID").agg(func.sum("passenger_count").alias("TotalNo_passengers")).sort('TotalNo_passengers',ascending=False).show()
```

```
from pyspark.sql import functions as func

df_revenue.groupBy("vendorID","tpep_pickup_datetime").agg(func.avg("revenue").alias("averageofTotal_amount")).sort('averageofTotal_amount',ascending=False).show()
```

```
df_revenue.filter('tpep_pickup_datetime = "01-01-2018 02:04"').display()
```

```
from pyspark.sql import functions as func

df.groupBy("payment_type").agg(func.sum("total_amount").alias("TotalAmount_bypayment")).orderBy('TotalAmount_bypayment',ascending=False).show()
```

```
from pyspark.sql.functions import col, substring
df_revenue = df_revenue.withColumn("record_date",substring(df_revenue['tpep_pickup_datetime'],0,10))
#below is the result
df_revenue.display()
```

```
df_revenue.printSchema()
```

```
from pyspark.sql import functions as func

df2=df_revenue.select(["record_date","vendorid","passenger_count","trip_distance"])\
.groupBy("record_date","vendorid","trip_distance").agg(func.sum("passenger_count").alias("totalpassenger")).sort('trip_distance',ascending=False).display()
```

```
%python
from pyspark.sql import functions as func

df_revenue_max=df_revenue.groupBy("PULocationID","DOLocationID").agg(func.sum("passenger_count").alias("totalpassenger")).orderBy("totalpassenger",ascending=False)
#df2.agg({"totalpassenger" : 'max'}).alias(max).display()
#result=df1.select([Max("totalPassenger")]).display()
```

```
df_revenue_max_value = df_revenue_max.select(func.max(df_revenue_max.totalpassenger)).collect()[0]
df_revenue_max.filter(df_revenue_max.totalpassenger ==df_revenue_max_value[0] ).display()
```
