```
from pyspark.sql import functions as f
```

```
df = spark.read.option("header",True).option("inferSchema",True).option("escape","\\").option("sep",",").option("multiLine",True).format("csv").load("/FileStore/tables/OI_MVP_Custs.csv")
df.display()
```

```
df1 = df.groupBy("name","address").count().where("count >1").sort("count",ascending = False)
df1.display()
```

```
df.filter('name = "GE Grid Solutions LLC"').display()
```

```
from pyspark.sql.functions import monotonically_increasing_id
df2 = df1.withColumn("MasterId",monotonically_increasing_id())
df2.display()
```

```
from pyspark.sql import Window
from pyspark.sql.functions import monotonically_increasing_id
df3 = df.withColumn('master_id', monotonically_increasing_id())\
.withColumn('master_id2', f.first('master_id').over(Window.partitionBy('name','address')))
df3.filter('name = "Graybar Electric Co., Inc."').display()
```

```
df3 = df3.drop("master_id")
df3 = df3.withColumnRenamed("master_id2","MasterId")
df3.display()
```
