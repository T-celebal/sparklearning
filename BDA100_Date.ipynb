```
from pyspark.sql import *
from pyspark.sql.types import *
from pyspark.sql.functions import *
```

```
cs = StructType([StructField("distinct_id",StringType(),True),StructField("event",StringType(),True),StructField("date",DateType(),True)])
df = spark.read.option("header",True).option("multiLine",True)\
.option("escape","\\").option("inferSchema",True).format("csv").load("/FileStore/tables/mixpanel_10day_sample.csv")
df2 = df.withColumn("date",to_date(df.date,"dd-MM-yyyy")) # to_date function is used to convert string into date
                                                           # to_date func takes date format from source column
df2.printSchema()
df2.display()
```

```
df3 = df2.select("distinct_id","date").distinct().groupBy("distinct_id","date").count() # it counts the grouped rows
df3.filter('date = "2020-10-02"').display()
df3.count() # this one counts all the rows present in the dataframe
```
